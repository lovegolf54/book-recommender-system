# Clean text

```python
import nltk
import string
nltk.download('stopwords')

from pythainlp.util import normalize
from pythainlp.tokenize import word_tokenize
from pythainlp.corpus import thai_stopwords
from nltk.corpus import stopwords
```

```python
def clean_text(text):
    text = normalize(text) #จัดการกับการพิมพ์ข้อความที่เรียงผิดหรือใช้ผิดอักษร
    text = text.lower() #แปลงตัวอักษรภาษาอังกฤษให้เป็นตัวพิมพ์เล็ก
    
    text = re.compile('\n').sub('',text) #กำจัดแท็ก\n
    text = re.compile('“').sub('',text) #กำจัด“
    text = re.compile('”').sub('',text) #กำจัด”
    text = re.compile('…').sub('',text) #กำจัด”
    text = re.compile('\xa0').sub('',text) #กำจัดแท็ก\xa0
    text = re.compile('chapter').sub('',text) #กำจัดคำว่า chapter
    text = re.compile('บท').sub('',text) #กำจัดคำว่า chapter
    text = re.compile('หน่วยการเรียนรู้').sub('',text) #กำจัดคำว่า chapter
    text = re.compile('['+string.punctuation+']').sub(' ',text) #กำจัดเครื่องหมายวรรคตอน !"#$%&'()*+,-./:;<=>?@[\]^_`{|}~
    

    return text
```

```python
def tokennize(text):
    thaistopwords = list(thai_stopwords()) #เก็บคำหยุดไว้ในตัวแปร
    stop_words = set(stopwords.words('english'))

    tokens = word_tokenize(text,keep_whitespace=False,engine="deepcut") #คำสั่งตัดคำ
    
    tokens = [i for i in tokens if i not in stop_words] #เก็บคำที่ไม่ตรงกับรายการคำหยุด
    tokens = [i for i in tokens if i not in thaistopwords] #เก็บคำที่ไม่ตรงกับรายการคำหยุด

    return tokens
```

```python
def count_stopword(text):
    thaistopwords = list(thai_stopwords()) #เก็บคำหยุดไว้ในตัวแปร
    stop_words = set(stopwords.words('english'))

    tokens = word_tokenize(text,keep_whitespace=False,engine="deepcut") #คำสั่งตัดคำ
    
    wordstop = [i for i in tokens if i in thaistopwords]
    
    return wordstop
```

```python
def identity_fun(text):
    return text
```

```python
tfidf_vectorizer = TfidfVectorizer(analyzer = 'word', #this is default
                                   tokenizer=identity_fun, #does no extra tokenizing
                                   preprocessor=identity_fun, #no extra preprocessor
                                   token_pattern=None,)

tfidf_vector= tfidf_vectorizer.fit_transform(df['tokens'])
tfidf_array = np.array(tfidf_vector.todense())

df1 = pd.DataFrame(tfidf_array,columns=tfidf_vectorizer.get_feature_names_out())
```

```python
df_binary = df1.applymap(lambda x: 1 if x > 0 else 0)
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import pandas as pd

# สร้าง DataFrame ที่มีค่า TF-IDF ของคำ
df_tfidf = df_binary

# กำหนดค่าสูงสุดของความน่าสนใจในกฎความสัมพันธ์
min_support = 0.1
min_threshold = 0.7

# หา Itemset ที่เกิดขึ้นบ่อยๆ ด้วย Algorithm Apriori
freq_itemset = apriori(df_tfidf, min_support=min_support, use_colnames=True)

# หากฎความสัมพันธ์ที่มีความน่าสนใจ
rules = association_rules(freq_itemset, metric="confidence", min_threshold=min_threshold)

# แสดงกฎความสัมพันธ์
print(rules)
```

# Building Model

```python
from sklearn.feature_selection import SelectPercentile, f_classif

selector = SelectPercentile(f_classif,percentile=10)
selector.fit(tfidf_vector,df['bal'])
```

```python
X = selector.transform(tfidf_vector).toarray()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

X_train, X_test, y_train, y_test = train_test_split(X,df['bal'],test_size=.3)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

from sklearn.metrics import confusion_matrix
```

```python
dtc = DecisionTreeClassifier(max_depth=5)
rfc = RandomForestClassifier(n_estimators=50, random_state=2)
mnb = MultinomialNB()
knc = KNeighborsClassifier()
svc = SVC(kernel='sigmoid',gamma=1.0)
```

```python
clfs = {
    'Decision Tree': dtc,
    'Random Forest': rfc,
    'Naive Bayes': mnb,
    'K-Nearest Neighbors': knc,
    'Support Vector Machines': svc
}
```

```python
def train_classifier(clf,X_train, X_test, y_train, y_test):
    clf.fit(X_train,y_train)
    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test,y_pred)
    precision = precision_score(y_test,y_pred,average='weighted')
    recall = recall_score(y_test,y_pred,average='weighted')
    f1 = f1_score(y_test,y_pred,average='weighted')
    
    cf_matrix = confusion_matrix(y_test, y_pred)

    return accuracy,precision,recall,f1,cf_matrix
```

```python
#train_classifier(svc,X_train, X_test, y_train, y_test)
```

```python
accuracy_scores = []
precision_scores = []
recall_scores = []
f1_scores = []
cf_matrixs=[]

for name,clf in clfs.items():
    current_accuracy,current_precision,current_recall,current_f1,current_cf = train_classifier(clf,X_train, X_test, y_train, y_test)
    print("For",name)
    print("Accuracy - ",current_accuracy)
    print("Precision - ",current_precision)
    print("Recall - ",current_recall)
    print("f1_score - ",current_f1)
    print("confusion_matrix \n ",current_cf)
    print("\n")
    
    
    accuracy_scores.append(current_accuracy)
    precision_scores.append(current_precision)
    recall_scores.append(current_recall)
    f1_scores.append(current_f1)
    cf_matrixs.append(current_cf)
```

```python
performance_df = pd.DataFrame({'Algorithm':clfs.keys(),
                               'Accuracy':accuracy_scores,
                               'Precision':precision_scores,
                               'Recall':recall_scores,
                               'F1_Score':f1_scores
                              })
#.sort_values('Accuracy',ascending=False)
```
